# Filtering datasets through boolean conditions

```
bool_cond = new_indexed_df.index.str.startswith('F') # create a boolean condition that checks if the index starts with 'A'
another_cond = new_indexed_df.turnout > 70 # check if turnout is greater than 7

new_indexed_df[bool_cond & another_cond] # apply the boolean condition to the indexed dataframe
```

```
cond_obama = new_indexed_df.winner == 'Obama' # check if winner is obama
new_indexed_df[cond_obama]
```

```
cond_romney = new_indexed_df.winner == 'Romney' # check if winner is romney
new_indexed_df[~cond_romney] # select the data that DO NOT meet the conditions 
```

# Query method

```
new_indexed_df.query('turnout > 70')
```

```
condition_one = 70
new_indexed_df.query('turnout > @condition_one') # the '@' symbol indicate that 'condition_one' is a global variable
```
